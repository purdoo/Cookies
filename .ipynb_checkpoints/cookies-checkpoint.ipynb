```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('cookies.csv',index_col=0,sep="\t")
# df.fillna('', inplace=True)
# Isolate cookie columns
cookie_cols = df.columns
```

```python
# Cookie rankings dict
cookie_ratings = {}
for col in df.columns:
    cookie_ratings[col] = df[col].mean()
cookie_ratings
    
```

```python
# Adding 'Rating Count' and 'User Average' cols
df['Rating Count'] = df.notna().sum(axis=1)
df['User Average'] = df.mean(axis=1)
df
```
